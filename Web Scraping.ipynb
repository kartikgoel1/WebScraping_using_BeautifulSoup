```python
# printing all the categories of books
from bs4 import BeautifulSoup
import requests
r = requests.get("http://books.toscrape.com/")
bs_obj = BeautifulSoup(r.text, 'html.parser')
lis = bs_obj.find(class_ = "side_categories").li.ul
categories = []
for i in lis.find_all('a'):
    categories.append(i.string)
for i in categories:
    i = i.replace('\n','')
    print(i.strip())
```

```python
# print the details of all books which are present on first 2 pages of this website.
#All details include - Title of the book, book page url, Price (in float, without any currency or extra symbol), and quantity in stock (in integer). Save all the details in a dataframe and print in the required format.
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']
basic_url = 'http://books.toscrape.com/catalogue/'
Title = []
Link = []
Price = []
qty = []
book_details = []
for i in allPages:
    r = requests.get(i)
    bs_obj = BeautifulSoup(r.text, 'html.parser')
    for i in bs_obj.find_all(class_ = 'product_pod'):
        Title.append(i.h3.a["title"].strip())
        Link.append(basic_url + i.h3.a["href"])
        r = requests.get(basic_url + i.h3.a["href"])
        bs_obj = BeautifulSoup(r.text, 'html.parser')
        price_string = bs_obj.find(class_ = 'price_color').string
        p = float(re.search('[\d.]+', price_string).group())
        Price.append(p)
        qty_string = bs_obj.find(class_ = 'instock availability').contents[-1].strip()
        q = int(re.search('\d+', qty_string).group())
        qty.append(q)
        book_details.append([i.h3.a["title"].strip(), basic_url + i.h3.a["href"], p, q]) 
       
        
        
df = pd.DataFrame(book_details, columns = column_names)

for i in range(len(df)):
    print(df['Title'][i],df['Link'][i],df['Price'][i],df['Quantity in Stock'][i])
```

```python
# printing all authors
from bs4 import BeautifulSoup
import requests
au = {}
base_url = "http://quotes.toscrape.com/page/"
for i in range(1,11):
    r = requests.get(base_url + str(i) + '/')
    bs_obj = BeautifulSoup(r.text, 'html.parser')
    for i in bs_obj.find_all(class_ = 'author'):
        au[i.string.strip()] = au.get(i.string.strip(), 0) + 1
sorted_dict = {k: au[k] for k in sorted(au)}
for author in sorted_dict:
    print(author.strip())
```
